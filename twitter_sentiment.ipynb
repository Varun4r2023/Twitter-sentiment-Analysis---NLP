```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load dataset: read as plain text (one line = one tweet)
with open("tweets_sample.csv", "r", encoding="utf-8") as f:
    lines = f.read().splitlines()

# Remove header if present
if lines[0].lower().strip() == "tweet":
    lines = lines[1:]

# Create DataFrame
df = pd.DataFrame(lines, columns=["tweet"])

# Sentiment analysis
def analyze_sentiment(tweet):
    polarity = TextBlob(str(tweet)).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["tweet"].apply(analyze_sentiment)

# Show results
print(df)

# Plot sentiment counts
df["Sentiment"].value_counts().plot(
    kind="bar",
    color=["green", "red", "blue"]
)
plt.title("Sentiment Analysis")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
